```python
from pydub import AudioSegment
import numpy as np
# Librosa for audio
import librosa
# And the display module for visualization
import librosa.display
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = []
labels = []
labellist = []
count=[]
pwd='../minor/ziggy/train_op_10clss/'
files = os.listdir(pwd)
for filename in files:
    labellist.append(filename)
for filename in labellist:
    print(filename)
    path=pwd+filename
    x=[]
    count.append(len(os.listdir(path)))
    for audio in os.listdir(path):
        try:
            wav = AudioSegment.from_wav(path+'/'+audio)
            wav=np.array(wav.get_array_of_samples()).astype(np.float32)
            S = librosa.feature.melspectrogram(wav, sr=16000, n_mels=128)
            log_S = librosa.power_to_db(S, ref=np.max)
            log_S1=np.zeros((128,32))
            m=log_S.shape[1]
            if m<=32:
                log_S1[:,:m]=log_S[:,:]
            else:
                log_S1[:,:]=log_S[:,:32]
            log_S1=log_S1.T
            log_S1=log_S1.reshape((32,128,1))
            x.append(log_S1)
        except:
            print(path+'/'+audio)
    X.append(x)
```

```python
for ix in range(len(count)):
    for ij in range(count[ix]):
        labels.append(labellist[ix])
labels=np.asarray(labels)
```

```python
for ix in range(len(X)):
    os.mkdir('out_datasets/'+labellist[ix])
    for ij in range(len(X[ix])):
        spec=X[ix][ij]
        np.savez('out_datasets/'+labellist[ix]+'/'+str(ij)+'.npz',spec)
        
np.savez('../out_datasets/audio/labels.npz',labels)
```
